## Функция для вычисления итоговой sMAPE

Ключевой метрикой для данной задачи будем считать метрику качества — sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»), которая вычисляется по формуле:


Отметим, что итоговое sMAPE состоит из двух слагаемых:


```python
def smape(target, prediction):
    numenator = abs(target - prediction)
    denominator =  (abs(target) + abs(prediction)) / 2
    return (1 / len(target)) * sum(numenator / denominator) * 100
```

```python
def total_smape(smape_r, smape_f):
    return 0.25*smape_r + 0.75*smape_f
```

## Обучение различных моделей и оценка их качества кросс-валидацией
