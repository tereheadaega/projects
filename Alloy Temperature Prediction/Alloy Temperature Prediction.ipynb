### Exploratory data analysis

#### Processing step description

Steel is processed in a metal ladle with a capacity of about 100 tons. In order for the ladle to withstand high temperatures, it is lined with refractory bricks from the inside. Molten steel is poured into a ladle and heated to the desired temperature with graphite electrodes. They are installed in the lid of the bucket.

Sulfur is removed from the alloy (desulfurization), the chemical composition is corrected by adding impurities, and samples are taken. Steel is alloyed - its composition is changed - by feeding pieces of alloy from a bunker for bulk materials or wire through a special tribe apparatus (English tribe, "mass").

Before introducing alloying additives for the first time, the temperature of the steel is measured and its chemical analysis is carried out. Then the temperature is raised for several minutes, alloying materials are added and the alloy is purged with an inert gas. Then it is stirred and measured again. This cycle is repeated until the target chemical composition and optimum melting temperature are reached.

Then the molten steel is sent to finish the metal or enters the continuous casting machine. From there, the finished product comes out in the form of slab blanks (English *slab*, “slab”).

### Selection of hyperparameters for the best model

The lowest value of the MAE metric was shown by the CatBoost model. After studying the importance of features, you can remove some. Next, an algorithm for finding the best hyperparameters will be used.

```python
features_train = features_train.drop(['Bulk 2','Bulk 13', 'Wire 7','Wire 4'], axis=1)
features_test = features_test.drop(['Bulk 2','Bulk 13', 'Wire 7','Wire 4'], axis=1)
```

```python
def objective_cat(trial, X, y):
    param = {
         "learning_rate": trial.suggest_float('learning_rate', 0.001, 0.04),
         "max_depth": trial.suggest_int('max_depth', 3, 12),
         "l2_leaf_reg": trial.suggest_discrete_uniform('l2_leaf_reg', 1.0, 4, 0.5),
         "min_child_samples": trial.suggest_categorical('min_child_samples', [1, 4, 8, 16, 32]),
        'loss_function': 'MAE'      
    }
    
    cv = KFold(n_splits=5, shuffle=True, random_state=RND)
    
    cv_predicts = np.empty(5)

    for idx, (train_idx, test_idx) in enumerate(cv.split(X, y)):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    
        model = CatBoostRegressor(verbose=False, **param, random_seed=RND)
    
        model.fit(X_train,
                  y_train,
                  eval_set=[(X_test, y_test)],
                  early_stopping_rounds=100,
                  use_best_model=True,
                  verbose=2)
        preds = model.predict(X_test)        
        cv_predicts[idx] = mean_absolute_error(y_test, preds)   
    
    
    
    return np.mean(cv_predicts)
```

```python
study_cat = optuna.create_study(direction="minimize", study_name="CatBoostRegressor")
func_cat = lambda trial: objective_cat(trial, features_train, target_train)

study_cat.optimize(func_cat, n_trials=10, show_progress_bar=True)
```

```python
print(f"\t Значение метрики (MAE): {study_cat.best_value:.2f}")
```

## Conclusion

This study was created for the metallurgical plant LLC "So temper steel". The purpose of the study: Optimization of production processes through the construction of a model that predicts the temperature of the steel. The target metric of this study is MAE, i.e. mean absolute error. At the beginning of the study, a plan was drawn up, consisting of three pillars: Conducting exploratory data analysis, Pre-processing and machine learning, and a final report.

Each of the stages was completed in full. A study was carried out in order to get acquainted with the technological process, an analysis of data from the customer was carried out. Seven tables were studied:
- `data_arc.csv` — electrode data;
- `data_bulk.csv` - data on the supply of bulk materials (volume);
- `data_bulk_time.csv` - data on the supply of bulk materials (time);
- `data_gas.csv` — data on alloy gas purge;
- `data_temp.csv` - temperature measurement results;
- `data_wire.csv` - data on wire materials (volume);
- `data_wire_time.csv` - data on wire materials (time).

It was determined that the data contained errors (Negative reactive power in the data_arc table and a temperature below the melting temperature in the ata_temp table). The distribution patterns of the used wire materials (table data_wire) and loose powders (table data_bulk) were studied. It was found that not all distributions correspond to normal ones. It is important to note that there are gaps in the data and some of the dimensions were missing for some keys, so when creating the pivot table, we used the inner join method.

At the next stage, the data was preprocessed and combined into a single table. At this stage, the key was excluded from the data_arc table, where the reactive power has a negative value, and the active power is greater than 3.5. When working with the data_arc table, new features were introduced - the total arc heating time for each key, the number of heatings for each key, and the ratio of active to reactive power. Next, the data_bulk and data_bulk_time tables were worked out. New features have been added - the amount of added bulk material and the time between adding the first and last bulk material. The data_gas table was taken unchanged. By analogy with the data_bulk and data_bulk_time tables, transformations were made in the data_wire and data_wire_time tables. The Wire 5 attribute was removed from the data_wire table, because in the entire set under study, it occurs once and cannot affect the result of the prediction. Based on similar considerations, the feature Bulk 8 was removed from the study. Further, keys were removed from the data_temp table, where the ladle temperature was lower than the melting temperature this is contrary to the studied technological process. A sign was also added - the primary temperature.

Then, after combining all the features into a single table, a heatmap was generated showing the degree of correlation between the features. It can be seen that between some features there is a high (absolute value above 0.9) degree of correlation. So the correlation coefficient between active and reactive power is 0.92, and between Bulk 12 and Bulk 7 the correlation coefficient is equal to one. As a result, it was decided to remove from consideration the following features with a high degree of correlation with other features: 'Reactive power', 'Bulk 7', 'Bulk 9', 'Wire 2', 'Wire 8', 'Wire 9'. The final step of this section was to split the set into training and test sets.

Next, the models will be trained on a set of 28 features: Power ratio, Total heating time, Number of heatings, Bulk 1, Bulk 2, Bulk 3, Bulk 4, Bulk 5, Bulk 6, Bulk 10, Bulk 11, Bulk 12, Bulk 13, Bulk 14, Bulk 15, Bulk count, Bulk additive time, Gas 1, Wire 1, Wire 3, Wire 4, Wire 6, Wire 7, Wire count, Wire time, First temperature, Number of temperature readings.


The next important block of research is the construction of predictive models and the choice of the best one. For this study, linear regression, decision tree, random forest, lgbmregressor and catboostregressor models were prepared and tested. Each of the models was trained using cross-validation and neg_mean_absolute_error scoring, the absolute value of which gives the target metric of this MAE study. All models showed a sufficiently good result and, with proper selection of hyperparameters, are able to obtain the required metric, however, the CatBoostRegressor model was chosen for further work. For this model, a diagram was drawn showing the importance of the features. From the data obtained, it can be established that the Bulk 2, Bulk 13, Wire 7, Wire 4 features have a low degree of significance in constructing a prediction and, therefore, can be removed from the study. Next, this model was run using the optuna algorithm to find the optimal hyperparameters. As a result of the work, the MAE metric on the training sample was 5.98. The parameters turned out to be optimal:
'learning_rate': 0.02892945666385754, 'max_depth': 11, 'l2_leaf_reg': 3.5, 'min_child_samples': 32 and random_state=80822.
The learning_rate parameter is responsible for the learning rate and is one of the main ones in models that use gradient descent.
The max_depth parameter is responsible for the maximum depth of the trees underlying the gradient boosting models.
The l2_leaf_reg parameter is the coefficient of the L2 regularization term.
The min_child_samples parameter is responsible for the minimum amount of data on the sheet. The random_state parameter is responsible for the reproducibility of the result. It was the search for the best combinations among these parameters that showed the best result of the metric.
To control the independence of the result from the existing sample, the model was tested on a test set and showed MAE 5.4, which may indicate the reproducibility of the result.

Separately, a check was carried out, the purpose of which is to check the adequacy of the spent resource for the resulting result. The comparison was carried out with a model predicting the average temperature in the training sample. The MAE of such a predictive model turned out to be higher than that of the trained model.

The most difficult in this study was the process of selecting hyperparameters that would allow us to obtain a quality metric lower than with the default parameters. The complexity was caused not only by the choice of intervals in which the search for optimal values is performed, but also by a combination of features that can give the desired result, so that the model training time does not exceed several hours. After several iterations of the search, the desired combination of parameters and intervals was found and the model showed the best result, while the search time for optimal parameters remained acceptable.


The result of this study was a CatBoostRegressor model with parameters {'learning_rate': 0.014411262246931934, 'max_depth': 6, 'l2_leaf_reg': 1.5, 'min_child_samples': 8} and fixed parameter random_state equal to 80822 trained on 28 signs, it is capable of predicting the temperature of the ladle heating with an error of MAE 5.5, which corresponds to the purpose of the study.
